```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
x = [1,2,3,4,5,6,7]
y = [50,52,25,36,15,62,45]
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x, y, "g+--") #see matplotlib documentation for format string i.e g+-- here
#can do same as above by writing the property names as well (shown below).
# plt.plot(x, y, color = "green", linewidth = 2, linestyle = "dashdot")
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label = "Max")
plt.plot(days, min_t, label = "Min")
plt.plot(days, avg_t, label = "Avg")

plt.legend(loc="best", shadow=True)
plt.grid()
```

# Bar chart

vertical bar chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit = [40,2,34,12]

xpos = np.arange(len(company))
plt.xticks(xpos, company)
plt.ylabel("revenue")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2, revenue, label="Revenue", width=0.4)
plt.bar(xpos+0.2, profit, label="Profit", width=0.4)
plt.legend()
```

horizontal bar chart

```python
plt.yticks(xpos, company)
plt.xlabel("revenue")
plt.title("US Tech Stocks")
plt.barh(xpos-0.2, revenue, label="Revenue", height=0.4)
plt.barh(xpos+0.2, profit, label="Profit", height=0.4)
plt.legend()
```
