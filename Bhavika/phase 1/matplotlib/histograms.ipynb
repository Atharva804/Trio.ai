```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bs_men=[113,80,90,150,149,88,93,115,135,80,72,88,130]
bs_women=[67,89,98,120,130,150,85,70,80,79,121,111,100]
```

```python
plt.xlabel("sugar range")
plt.ylabel("no. of patients")
plt.title("blood sugar chart")
plt.hist ([bs_men,bs_women],bins=[80,100,125,150],rwidth=0.95,color=['green','red'],label=['men','women'],orientation='horizontal')
plt.legend()
```
