```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max=[32,51,52,48,47,49,46]
min=[43,42,40,44,33,35,37]
avg=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("days")
plt.ylabel("temp")
plt.title("weather")
plt.plot(days,max ,label="max")
plt.plot(days,min ,label="min")
plt.plot(days,avg , label ="avg")
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
```
