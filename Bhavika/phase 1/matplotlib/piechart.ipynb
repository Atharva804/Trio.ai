```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
value=[1400,600,300,410,250]
exp_labels=['home rent','food','phone bill','car','others']
```

```python
plt.axis="equal"
plt.pie(value,labels=exp_labels,radius=1.5,autopct='%0.2f%%',explode=[0.1,0.1,0.0,0.2,0.3],startangle=180)
plt.savefig('piechart.png', bbox_inches="tight",pad_inches=2)
```
