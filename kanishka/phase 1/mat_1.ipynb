```python
import matplotlib.pyplot as plt
%matplotlib inline
days = [1,2,3,4,5,6,7]
max_t = [50,48,46,44,42,40,38]
min_t = [37,36,35,34,33,32,31]
avg_t = [44,45,40,42,53,42,41]
plt.xlabel('Day')
plt.ylabel('Temoerature')
plt.title('Weather')
plt.plot(days,max_t,label ="Max")
plt.plot(days,min_t,label ="Min")
plt.plot(days,avg_t,label ="Avg")
plt.legend(shadow=True,fontsize="large")
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
company=["google","amazon","microsoft","facebook"]
revenue = [56,43,65,32]
profit=[40,32,21,11]
ypos = np.arange(len(company))
ypos
plt.xticks(ypos,company)
plt.bar(ypos,revenue)
plt.bar(ypos,revenue)
plt.title("us tech stocks")
```

```python
company=["google","amazon","microsoft","facebook"]
revenue = [56,43,65,32]
profit=[40,32,21,11]
ypos = np.arange(len(company))
xpos = np.arange(len(company))
xpos
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.bar(xpos-0.2,revenue,label = "Revenue")
plt.bar(xpos+0.2,profit,label = "Profit")
plt.legend()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
exp_vals = [1500,400,1200,350,500]
exp_labels = ["Rent","Food","Bill","Bike","Other"]
plt.pie(exp_vals,labels = exp_labels,radius = 1.5,autopct='%0.1f%%',shadow=True,explode=[0,0,1,0,0],startangle=180)
plt.show()
plt.savefig("piechart.png",bbox_inches="tight",pad_inches=1)
```
